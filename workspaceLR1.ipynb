```python
def f_sin_x(x, k = 1.0, x0=0.0, A=1.0, y0=0.0):
    return A * np.sin(k * (x - x0)) + y0
def f_cosh_x(x, k = 1.0, x0=0.0, A=1.0, y0=0.0):
    return A * np.cosh(k * (x - x0)) + y0
def f_tan_x(x, k = 1.0, x0=0.0, A=1.0, y0=0.0):
    return A * np.tan(k * (x - x0)) + y0
```
