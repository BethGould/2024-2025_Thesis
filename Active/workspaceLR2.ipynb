```python
#--LIBRARIES--------

import eelib
from eelib import pi
import numpy as np
import pickle
```

# $\mu$ Grid

```python
#--PARAMETERS--------

# note that k, B, and R are percents here, mu is not as its scale is unknown
dk = 0.5
R  = 1.0
B  = 0.5
mu = 1.0e-8

n = 11

r_r = (0.9, 0.9002)
b_r  = (0.5, 0.506) # B here is B*R, since varying R will alter k, which I don't want
dk_r = (0.1, 0.9)
mu_r = (-9.5, -7.5) #1.0e-8 to 1.0e-5
```

mu has 11 points from 3.1622776601683795e-10 to 3.162277660168379e-08.
dk has 11 points from 0.1 to 0.9.
B has 11 points from 0.5 to 0.506.
R has 11 points from 0.9 to 0.9002.

```python
#--CODE--

#make loop, grid, or system
gridl = eelib.grid_bvp.grid_BVP(R, B, dk, mu)
#gridl.makeMCPoints(mu = mu_r, dk = dk_r, B = b_r, R = r_r, num = n)
#gridl.mcBVP()
gridl.makeGridPoints(mu = mu_r, dk = dk_r, B = b_r, R = r_r, num = [n, n, n, n])
gridl.gridBVP()
```

```python
# since this takes a while to run and is used for a while, it is best to save a copy of the data
file = open('grid026b', 'ab')
pickle.dump(gridl, file)
file.close()
```
