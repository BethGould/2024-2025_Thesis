```python

#--LIBRARIES--------

#numerics
import numpy as np
import eelib

import pickle

#--PARAMETERS--------

# set parameters
# note that k, B, and R are percents here, mu is not as its scale is unknown
k = 0.5
R = 0.8
B = 0.5
mu = 1.0e-6

#n_k = 2
#n_R = 2
#n_B = 2
#n_a = 2
#n_m = 3
n_g = 9

#a_min = 0.5
#a_max = 3.0
#m_max = 3.0e-6
#m_min = 1.0e-7
#k_min = 0.5
#k_max = 0.7
#b_min = 0.5
#b_max = 2

#pr = 200
#n = 20

i = 0

filename_start = "loop"
filename2_start = "out"

mu_v = [1e-7, 1e-6, 3e-6]

#--CODE------------

#loading script; be careful of python version
#file = open('grid001', 'rb')    
#gridl = pickle.load(file)
#file.close()

for mu in mu_v:
    i = i + 1
    filename = f"{filename_start}{i:03}"
    filename2 = f"{filename2_start}{i:03}"

    #make loop, grid, or system
    loopl = eelib.deriv_grid(R, B, k, mu, grid_size=n_g)

    # solve the ivp or bvp
    loopl.derivGrid()

    loopl.calc_period(filename=filename2)

    file = open(filename, 'ab')
    pickle.dump(loopl, file)
    file.close()
```
