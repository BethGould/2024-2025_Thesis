```python
#--LIBRARIES--------

import eelib
from eelib import pi
import numpy as np
import pickle
```

# B R Grid

```python
#--PARAMETERS--------

# note that k, B, and R are percents here, mu is not as its scale is unknown
dk = 0.45
R  = 1.0
B  = 0.5
mu = 1.5e-8

n_b = 100
n_r = 100

r_r = (0.9, 0.9002)
b_r  = (0.5, 0.6) # B here is B*R, since varying R will alter k, which I don't want
dk_r = (0.1, 0.9)
mu_r = (-9.5, -7.5) #1.0e-8 to 1.0e-5
```

B has 100 points from 0.5 to 0.6.
R has 100 points from 0.9 to 0.9002.

```python
#--CODE--

#make loop, grid, or system
gridl = eelib.grid_bvp.grid_BVP(R, B, dk, mu)
#gridl.makeMCPoints(mu = mu_r, dk = dk_r, B = b_r, R = r_r, num = n)
#gridl.mcBVP()
gridl.makeGridPoints(B = b_r, R = r_r, num = [n_b, n_r])
gridl.gridBVP()
```

```python
# since this takes a while to run and is used for a while, it is best to save a copy of the data
file = open('grid030b', 'ab')
pickle.dump(gridl, file)
file.close()
```
