```python
#--LIBRARIES--------
import eelib
from eelib import pi
import numpy as np
import pickle
```

# Run 1

```python
#--PARAMETERS--------

# note that k, B, and R are percents here, mu is not as its scale is unknown
dk = 0.45
R  = 1.0
B  = 0.5

n_g = 1000
n_mu = 20

r_r = (0.57, 1.0)
b_r  = (0.11, 1.0)
dk_r = (0.1, 0.9)
mu_r = (5e-10, 5e-8)

#k0_r #later ....
```

```python
# find random mu values
rng = np.random.default_rng()
mu_arr = []
for i in range(n_mu):
    mu_arr.append(mu_r[0] + rng.random()*(mu_r[1]-mu_r[0]))
```

```python
# for each random mu value, make a mc object with parameters
for i, mu in enumerate(mu_arr):
    gridl = eelib.grid_bvp.grid_BVP(R, B, dk, mu)
    gridl.makeMCPoints(dk = dk_r, B = b_r, R = r_r, num = n_g)
    gridl.mcBVP()

    filename = "grid_"+str(i)
    file = open(filename, 'ab')
    pickle.dump(gridl, file)
    file.close()
```
